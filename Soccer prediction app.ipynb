```python
grouped_matches = matches.groupby("team") #improving precision with rolling averages. rolling averages/team performance. one datframe 
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date") #sorted by date
    rolling_stats = group[cols].rolling(3, closed='left').mean() #set of col & compute rolling averages 3 week assign to week 3
    group[new_cols] = rolling_stats #assign them back
    group = group.dropna(subset=new_cols) #drop any rows with missing values
    return group 
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))

matches_rolling #apply one function to each team in the data and apply a function to compute the rolling averages

```

```python
matches_rolling = matches_rolling.droplevel('team')

matches_rolling #drop extra index level

```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors): 
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True) #merge in colums fromrolling games from ondex.
```
